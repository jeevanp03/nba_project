```python
import mysql.connector as connection
import pandas as pd
import numpy as np
import os
try:
    mydb = connection.connect(host= os.environ.get("DB_HOST"), database = os.environ.get("DB_NAME"),user= os.environ.get("DB_USER"), passwd= os.environ.get("DB_PASSWORD"),use_pure=True)
    query = "Select * from nba_data;"
    nba_df = pd.read_sql(query,mydb)
    mydb.close() #close the connection
except Exception as e:
    mydb.close()
    print(str(e))
```

```python
clean_df = nba_df.copy()
clean_df.sort_values(by = ['PLAYER_NAME','season_id','NBA_FANTASY_PTS_RANK','PTS_RANK',], inplace = True)
```

```python
names = pd.unique(clean_df[['PLAYER_NAME']].values.ravel())
player_id = pd.unique(clean_df[['id']].values.ravel())
```

```python
new_df = pd.DataFrame(player_id, columns =["id"])
new_df['PLAYER_NAMES'] = names
print(names)
```

```python
print(new_df)
```

```python
all_names = clean_df["PLAYER_NAME"].tolist()
print(all_names)
```

```python
length = len(all_names)
print(length)
print(len(names))
```

```python
i = 0
count = 1
num = []
while i < length-1:
    if all_names[i] == all_names[i+1]:
        count+=1
        print(count)
    else:
        num.insert(i, count)
        count = 1
        print(count)
    i +=1
if all_names[length-2] == all_names[length-1] and all_names[length-3] == all_names[length-2]:
    count +=1
    print(count)
    num.insert(i-1, count)
elif all_names[length-2] == all_names[length-1]:
    count = 1
    count +=1
    print(count)
    num.insert(i, count)
else:
    count = 1
    print(count)
    num.insert(i, count)
```

```python
print(num)
```

```python
print(len(num))
```

```python
new_df['count'] = num
print(new_df)
```

```python
clean_df = pd.merge(clean_df, new_df, how = 'right', on = 'id')
```

```python
print(clean_df)
```

```python
clean_df = clean_df.drop('PLAYER_NAMES', 1)
clean_df = clean_df.drop('WNBA_FANTASY_PTS_RANK', 1)
clean_df = clean_df.drop('WNBA_FANTASY_PTS', 1)
print(clean_df)
```

```python
from sqlalchemy import create_engine
from pandas.io import sql
import pymysql
```

```python
import mysql.connector
mydb = mysql.connector.connect(
  host=os.environ.get("DB_HOST"),
  user=os.environ.get("DB_USER"),
  password=os.environ.get("DB_PASSWORD"),
  database=os.environ.get("DB_NAME")
)

mycursor = mydb.cursor()

sql = "DROP TABLE IF EXISTS nba_clean"

mycursor.execute(sql)
```

```python
# Credentials to database connection
hostname=os.environ.get("DB_HOST")
dbname=os.environ.get("DB_NAME")
uname=os.environ.get("DB_USER")
pwd=os.environ.get("DB_PASSWORD")
```

```python
# Create SQLAlchemy engine to connect to MySQL Database
engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}"
				.format(host=hostname, db=dbname, user=uname, pw=pwd))
```

```python
clean_df.to_sql('nba_clean', engine, index=False)
```
