```python
season_list = [
    '2009-10',
    '2010-11',
    '2011-12',
    '2012-13',
    '2013-14',
    '2014-15',
    '2015-16',
    '2016-17',
    '2017-18',
    '2018-19',
    '2019-20',
    '2020-21',
    '2021-22',]
```

```python
dfs = []
```

```python
# for loop to loop over seasons
for season_id in season_list:
    player_info_url = 'https://www.basketball-reference.com/leagues/NBA_'+season_id+'_per_game.html'
    source = requests.get(url, headers = headers)
    soup = BeautifulSoup(source.content,'lxml')
    # pulling just the data we want
    header = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    header = header[1:]
    rows = soup.findAll('tr')[1:]
    player_stats =[[td.getText() for td in rows[i].findAll('td')]for i in range(len(rows))]
    df = pd.DataFrame(player_stats, columns = header)
    df['season_id'] = season_id
    print(season_id)
    dfs.append(df)
```

```python
final_df = pd.concat(dfs)
```

```python
final_df.sort_values(by = ['Player','season_id','PTS'], inplace = True)
final_df
```

```python
final_df = final_df.dropna()
final_df
```

```python
final_df = pd.merge(final_df, first_df, how = 'right', on = 'Player')
```
